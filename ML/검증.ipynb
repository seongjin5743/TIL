# 검증 세트
- 훈련 세트에서 검증 세트를 약 20퍼 정도 주는 방식.
- 테스트하고 싶은 모델을 훈련하고 검증해 가장 좋은 것을 고른다.

```python
import pandas as pd
wine = pd.read_csv('https://bit.ly/wine_csv_data')
wine
```

```python
data = wine[['alcohol', 'sugar', 'pH']]
target = wine[['class']]
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(data, target)
```

```python
sub_input, val_input, sub_target, val_target = train_test_split(train_input, train_target)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(max_depth = 50)
dt.fit(sub_input, sub_target)
print(dt.score(sub_input, sub_target))
print(dt.score(val_input, val_target))
```

```python
print(dt.score(test_input, test_target))
```

# 교차 검증
- 평가하는 과정을 여러 번 반복해 점수를 편균을 내서 최종 점수를 얻는다.

```python
from sklearn.model_selection import cross_validate

scores = cross_validate(dt, train_input, train_target)
print(scores)
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    'min_impurity_decrease' : [0.0001, 0.0002, 0.0003, 0.0004]
}
gs = GridSearchCV(dt, params, n_jobs = -1)
gs.fit(train_input, train_target)
```

```python
gs.best_estimator_
```
