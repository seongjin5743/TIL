# 훈련, 테스트

```python
import pandas as pd

df = pd.read_csv('data/fish.csv')

con = df['Species'].isin(['Bream', 'Smelt'])
lw_df = df.loc[con, ['Length2', 'Weight']]
fish_data = lw_df.values.tolist()
fish_target = [1] * 35 + [0] * 14
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
```

```python
train_input = fish_data[:35]
train_target = fish_target[:35]

test_input = fish_data[35:]
test_target = fish_target[35:]
```

```python
kn.fit(train_input, train_target)
```

```python
kn.score(test_input, test_target)
```

```python
import numpy as np

input_arr = np.array(fish_data)
target_arr = np.array(fish_target)

np.random.seed(42)
index = np.arange(49)
np.random.shuffle(index)
```

```python
input_arr[[1, 3]]
```

```python
train_input = input_arr[index[:35]]
train_target = target_arr[index[:35]]
```

```python
test_input = input_arr[index[35:]]
test_target = target_arr[index[35:]]
```

```python
import matplotlib.pyplot as plt

plt.scatter(train_input[:, 0], train_input[:, 1])
plt.scatter(test_input[:, 0], test_input[:, 1])
```

```python
kn.fit(train_input, train_target)
```

```python
kn.score(test_input, test_target)
```

```python
kn.predict(test_input)
```

## iris

```python
# setosa, versicolor
# petal_length, petal_width

import seaborn as sns
iris = sns.load_dataset('iris')

con = iris['species'].isin(['setosa', 'versicolor'])
iris_df = iris.loc[con, ['petal_length', 'petal_width']]

iris_data = iris_df.values
```

```python
setosa_count = iris.loc[iris['species'] == 'setosa'].shape[0]
versicolor_count = iris.loc[iris['species'] == 'versicolor'].shape[0]
print(setosa_count)
```

```python
import numpy as np

iris_target = np.array([1] * setosa_count + [0] * versicolor_count)
```

```python
import random

index = list(range(setosa_count + versicolor_count))
random.shuffle(index)
```

```python
train_input = iris_data[index[:70]]
train_target = iris_target[index[:70]]
test_input = iris_data[index[70:]]
test_target = iris_target[index[70:]]
```

```python
plt.scatter(train_input[:, 0], train_input[:, 1])
plt.scatter(test_input[:, 0], test_input[:, 1])
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
```

```python
kn.fit(train_input, train_target)
```

```python
kn.score(test_input, test_target)
```

```python
kn.predict(test_input)
```
