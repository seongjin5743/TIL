# K 최근접 이웃

```python
import pandas as pd

df = pd.read_csv('data/fish.csv')
df.head()
```

```python
bream_df = df.loc[df['Species'] == 'Bream']
bream_l = bream_df['Length2'].tolist()
bream_w = bream_df['Weight'].tolist()
```

```python
import matplotlib.pyplot as plt

# plot: 선, bar: 막대, scatter: 산점도
plt.scatter(bream_l, bream_w)
```

```python
smelt_df = df.loc[df['Species'] == 'Smelt']
smelt_l = smelt_df['Length2'].tolist()
smelt_w = smelt_df['Weight'].tolist()
```

```python
import matplotlib.pyplot as plt

# plot: 선, bar: 막대, scatter: 산점도
plt.scatter(smelt_l, smelt_w)
plt.scatter(bream_l, bream_w)
```

```python
length = bream_l + smelt_l
weight = bream_w + smelt_w

fish_data = [[l, w] for l, w in zip(length, weight)]
```

```python
fish_target = [1] * len(bream_w) + [0] * len(smelt_w)
```

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)
```

```python
kn.score(fish_data, fish_target)
```

```python
kn.predict([[30, 600]])
```

```python
kn.predict([[10, 20]])
```

```python
kn49 = KNeighborsClassifier(n_neighbors = 49)
kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
```
